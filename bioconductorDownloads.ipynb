# Bioconductor downloads

Le nombre de téléchargement des paquets software depuis Bioconductor est fourni sur http://bioconductor.org/packages/stats/  

le but est de récupérer ces données et de mettre par la suite un script qui va récupérer de manière automatique les données fournies au fur et à mesure du temps. Le lien au dessus surveille le téléchargements des paquets Bioconductor http://bioconductor.org/ au cours des 12 derniers mois. Le nombre déclaré à côté de chaque nom de paquet est le nombre d'adresses IP distinctes qui ont "frappé" le paquet au cours des 12 derniers mois.

```python
__author__ = 'narjissetabout'

from bs4 import BeautifulSoup
import urllib2

#downBioc = open('DownBioconductor.csv', 'w') #on peut decommenter pour sauvegarder les résultats
url = urllib2.urlopen('http://bioconductor.org/packages/stats/')  #je recupère le lien 
html = url.read()
soup = BeautifulSoup(html)
#count = 0
result = []
for k in soup.find_all('tr', "totaldlperpkg"): #les paquets avec downloads <tr><td></td><td><a>paquet; down</a></td></tr>
    for i in k.find_all('a'):
        if i.string is not None:
            href = i.get("href")
            splt1 = i.string.encode('utf-8').split("\xc2\xa0(")
            splt2 =splt1[1].split(")")
            #print >>downBioc, splt1[0]+","+splt2[0] # si on souhaite sauvegarder les résultats dans downBioc
            print(splt1[0]+","+splt2[0])
        for j in i.find_all('b'): #les paquets avec downloads <a><b>paquet</b> ;down</a>
            splt3 = j.next_sibling.encode('utf-8').split("\xc2\xa0(")
            splt4 = splt3[1].split(")")
            print(j.string+","+splt4[0])
            #print >>downBioc, j.string+","+splt4[0] #si on souhaite sauvegarder les résultats dans downBioc

#downBioc.close()
```
